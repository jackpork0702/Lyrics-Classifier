```python
from lyrics_package import data_preparation as dp
from lyrics_package import visiualization as vs
from lyrics_package import model_building as mb
```

```python
df =dp.prepare_data("C:/Users/jackpork0702/Desktop/lyrics_kaun/lyrics.csv")
df=dp.select_year(df,2008,2018)

Xr,Xh,Xm,Xc,Xw,Y=dp.df_to_array(df)
Xv,tfidfv = dp.vectorizer(Xw)


X_train, X_test, Y_train, Y_test=mb.split_train_test(Xv, Y)
```

## 08~18 songs Radom Forest Outcome

```python
import matplotlib.pyplot as plt
gen = ['rock', 'hip hop', 'metal', 'country']
rf_score,rf_cnf_matrix=mb.rf_model(X_train, X_test, Y_train, Y_test, gen)
print(rf_score)
print(rf_cnf_matrix)
```

```python
vs.plot_confusion_matrix(rf_cnf_matrix, gen,
                          normalize=True,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues)
```

## 08~18 songs Adaboost Outcome

```python
ada_score,ada_cnf_matrix=mb.ada_model(X_train, X_test, Y_train, Y_test, gen)
print(ada_score)
print(ada_cnf_matrix)
```
